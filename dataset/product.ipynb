```python
am_df = pd.read_csv("amazon.csv")
```

```python
Data = am_df.copy()
```

```python
# Removing Unwanted Columns:
Data.drop(columns=['review_id', 'user_id', 'user_name', 'img_link','product_link',
                   'discounted_price','discount_percentage','review_title','review_content'], inplace=True)
```

```python
Data['actual_price'] = Data['actual_price'].str.replace("₹",'')
Data['actual_price'] = Data['actual_price'].str.replace(",",'')
Data['actual_price'] = Data['actual_price'].astype('float64')
```

```python
Data['rating_count'] = Data['rating_count'].str.replace(",",'').astype('float64')
Data['category'].unique()
```

```python
# Split the 'category' column into multiple columns
split_columns = Data['category'].str.split('|', expand=True)

# Assign column names
split_columns.columns = ['Level1', 'Level2', 'Level3', 'Level4', 'Level5','Level6','Level7']

# Display the resulting DataFrame
split_columns.head(10)
```

```python
split_columns.isnull().sum()
```

```python
split_columns=split_columns.drop(columns= ['Level3','Level4','Level7','Level6','Level5'])
split_columns.columns = ['Category', 'sub_category']
split_columns['item'] = Data['category'].str.rsplit('|').str[-1]
split_columns
```

```python
# Final Dataset form
Data = Data.drop("category",axis = 1)
Data = Data.join(split_columns)
Data.info()
```
